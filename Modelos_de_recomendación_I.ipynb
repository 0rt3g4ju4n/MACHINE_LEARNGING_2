# Chartbook clock

```python
df = pd.read_csv('https://raw.githubusercontent.com/prnvg/Charles-Book-Club/refs/heads/master/CBC.csv', index_col=0)
```

```python
df.shape
```

```python
df.head()
```

```python
df.columns
```

```python
df_books = df[['ChildBks', 'YouthBks',
       'CookBks', 'DoItYBks', 'RefBks', 'ArtBks', 'GeoBks', 'ItalCook',
       'ItalAtlas', 'ItalArt', 'Florence']]
```

* $(ChildBks, YouthBks) \to (CookBks)$
* $(ChildBks, RefBks) \to (CookBks)$
* $(DoItYBks, ChildBks) \to (CookBks)$

```python
df_model = pd.DataFrame(np.where(df_books>0, 1, 0), columns = df_books.columns, index=df_books.index)
```

```python
#df_books = (df_books > 0).astype('int')
```

```python
df_model
```

```python
items_books = apriori(df_model, min_support=0.1, use_colnames=True)
```

```python
items_books
```

```python
association_rules(items_books, min_threshold=0.2).sort_values(by = ['confidence', 'zhangs_metric'], ascending=[False, False])
```

(ChildBks,YouthBks)→(CookBks)

(ChildBks,RefBks)→(CookBks)

(DoItYBks,ChildBks)→(CookBks)

* (ChildBks, RefBks)	(CookBks)
* (GeogBks, CookBks)	(ChildBks)
* (ChildBks, CookBks)	(DoItYBks)

```python
items_books = fpgrowth(df_model, min_support=0.05, use_colnames=True)
```

```python
items_books
```

```python
association_rules(items_books, min_threshold=0.4, metric='confidence').sort_values(by = ['confidence', 'zhangs_metric'], ascending=[False, False])
```
