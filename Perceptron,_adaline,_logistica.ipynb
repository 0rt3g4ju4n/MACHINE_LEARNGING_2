```python
from sklearn.datasets import load_digits
from sklearn.linear_model import Perceptron
```

```python
# prompt: generame un conjunto de puntos con la librería sklearn que sea linealmente separable

from sklearn.datasets import make_classification

# Generar un conjunto de datos linealmente separable
X, y = make_classification(n_features=2,
                           n_redundant=0,
                           n_informative=2,
                           random_state=1,
                           n_clusters_per_class=1,
                           class_sep=2)
```

```python
# prompt: realiza una visualización del conjunto de puntos generado, distinguiendo el conjunto de entrenamiento y el conjunto de pruebas

import matplotlib.pyplot as plt

plt.scatter(X[:, 0], X[:, 1], c=y, cmap='viridis')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Conjunto de Datos Linealmente Separable')
plt.show()
```

```python
clf = Perceptron(tol=1e-3, random_state=0)
clf.fit(X, y)
```

```python
clf.score(X, y)
```

```python
# prompt: dibuja el conjunto de datos y la recta aproximada del perceptron

import matplotlib.pyplot as plt
import numpy as np

# Crear una malla de puntos para representar la recta de decisión
xx, yy = np.meshgrid(np.linspace(X[:, 0].min() - 1, X[:, 0].max() + 1, 100),
                     np.linspace(X[:, 1].min() - 1, X[:, 1].max() + 1, 100))

# Predecir las etiquetas para cada punto de la malla
Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

# Dibujar la recta de decisión
plt.contourf(xx, yy, Z, alpha=0.4)

# Dibujar los puntos de datos
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='viridis')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Conjunto de Datos con Recta de Decisión del Perceptrón')
plt.show()
```

```python
from sklearn.linear_model import LogisticRegression

clf = LogisticRegression(random_state=0)
clf.fit(X, y)
```

```python
clf.score(X, y)
```

```python
# prompt: dibuja el conjunto de datos con la predicción realizada mediante la regresión logistica

import matplotlib.pyplot as plt
import numpy as np
# Crear una malla de puntos para representar la recta de decisión
xx, yy = np.meshgrid(np.linspace(X[:, 0].min() - 1, X[:, 0].max() + 1, 100),
                     np.linspace(X[:, 1].min() - 1, X[:, 1].max() + 1, 100))

# Predecir las etiquetas para cada punto de la malla
Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

# Dibujar la recta de decisión
plt.contourf(xx, yy, Z, alpha=0.4)

# Dibujar los puntos de datos
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='viridis')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Conjunto de Datos con Recta de Decisión de la Regresión Logística')
plt.show()
```

```python
# prompt: realiza el modelo de clasificación binaria para el conjunto de datos con adaline

import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import SGDClassifier

# Crear un modelo Adaline utilizando SGDClassifier con la función de pérdida 'perceptron'
clf = SGDClassifier(loss='perceptron', random_state=0)

# Entrenar el modelo con los datos
clf.fit(X, y)

# Calcular la precisión del modelo
clf.score(X, y)

# Crear una malla de puntos para representar la recta de decisión
xx, yy = np.meshgrid(np.linspace(X[:, 0].min() - 1, X[:, 0].max() + 1, 100),
                     np.linspace(X[:, 1].min() - 1, X[:, 1].max() + 1, 100))

# Predecir las etiquetas para cada punto de la malla
Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

# Dibujar la recta de decisión
plt.contourf(xx, yy, Z, alpha=0.4)

# Dibujar los puntos de datos
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='viridis')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Conjunto de Datos con Recta de Decisión de Adaline')
plt.show()
```
